<h2>Data Scarping From Ever Care Hospital</h2>

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

urls = [
    "https://www.evercarebd.com/en/dhaka/specialities/counselling-center",
    "https://www.evercarebd.com/en/dhaka/specialities/child-development-centre",
    "https://www.evercarebd.com/en/dhaka/specialities/cardiology",
    "https://www.evercarebd.com/en/dhaka/specialities/cardiothoracic-vascular-surgery",
    "https://www.evercarebd.com/en/dhaka/specialities/cardiothoracic-anaesthesia",
    "https://www.evercarebd.com/en/dhaka/specialities/anaesthesia-and-pain-medicine",
    "https://www.evercarebd.com/en/dhaka/specialities/accident-emergency",
    "https://www.evercarebd.com/en/dhaka/specialities/diabetology-endocrinology",
    "https://www.evercarebd.com/en/dhaka/specialities/dermatology-venereology",
    "https://www.evercarebd.com/en/dhaka/specialities/dental-maxillofacial-surgery",
    "https://www.evercarebd.com/en/dhaka/specialities/critical-care-units",
    "https://www.evercarebd.com/en/dhaka/specialities/general-laparoscopic-surgery",
    "https://www.evercarebd.com/en/dhaka/specialities/gastroenterology-hepatology",
    "https://www.evercarebd.com/en/dhaka/specialities/fertility-centre",
    "https://www.evercarebd.com/en/dhaka/specialities/ent-head-neck-surgery",
    "https://www.evercarebd.com/en/dhaka/specialities/dietetics-nutrition",
    "https://www.evercarebd.com/en/dhaka/specialities/lab-medicine",
    "https://www.evercarebd.com/en/dhaka/specialities/kidney-transplant-program",
    "https://www.evercarebd.com/en/dhaka/specialities/joint-care-wellness-centre",
    "https://www.evercarebd.com/en/dhaka/specialities/internal-medicine",
    "https://www.evercarebd.com/en/dhaka/specialities/hepatobiliary-pancreatic-and-liver-transplant",
    "https://www.evercarebd.com/en/dhaka/specialities/nuclear-medicine",
    "https://www.evercarebd.com/en/dhaka/specialities/neurosurgery",
    "https://www.evercarebd.com/en/dhaka/specialities/neurology",
    "https://www.evercarebd.com/en/dhaka/specialities/nephrology",
    "https://www.evercarebd.com/en/dhaka/specialities/lithotripsy",
    "https://www.evercarebd.com/en/dhaka/specialities/thoracic-surgery",
    "https://www.evercarebd.com/en/dhaka/specialities/rheumatology",
    "https://www.evercarebd.com/en/dhaka/specialities/respiratory-medicine",
    "https://www.evercarebd.com/en/dhaka/specialities/psychiatry",
    "https://www.evercarebd.com/en/dhaka/specialities/plastic-reconstructive-cosmetic-surgery",
    "https://www.evercarebd.com/en/dhaka/specialities/physical-medicine-rehabilitation",
    "https://www.evercarebd.com/en/dhaka/specialities/hematology-stem-cell-transplant",
    "https://www.evercarebd.com/en/dhaka/specialities/urology",
    "https://www.evercarebd.com/en/dhaka/specialities/transfusion-medicine",
    
]

# Create DataFrame
df = pd.DataFrame(columns=[
    "Doctor Name",
    "Designation",
    "Department",
    "Hospital",
    "Address"
])

for url in urls:
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")

    # Get department name
    dept = soup.find("h1")
    department_name = dept.text.strip() if dept else "Unknown"

    # Target the correct class
    cards = soup.find_all("div", class_="flex-1 items-center justify-center rounded-xl border border-sky-400 text-center")

    for card in cards:
        text_lines = card.get_text(separator="\n").split("\n")
        text_lines = [t.strip() for t in text_lines if t.strip()]

        # Extract fields safely
        name = text_lines[0] if len(text_lines) > 0 else ""
        designation = text_lines[1] if len(text_lines) > 1 else ""

        row_data = [
            name,
            designation,
            department_name,
            "Evercare Hospital Dhaka",
            "Plot #81, Block-E, Bashundhara R/A, Dhaka"
        ]

        length = len(df)
        df.loc[length] = row_data

    time.sleep(2)

# Print table
print(df.to_string(index=False))
```

```python
df.to_csv("doctors.csv", index=False)
```
