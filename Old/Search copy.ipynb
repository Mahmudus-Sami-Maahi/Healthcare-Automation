```python
import pandas as pd
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq

# 1. Load CSV Data
def load_csv(file_path):
    df = pd.read_csv(file_path)
    return df.to_string(index=False)  # Convert table to text for LLM

# 2. Initialize LLM
llm = ChatGroq(
    model="openai/gpt-oss-120b",
    temperature=0.2
)

# 3. Prompt Template
template = """
You are a helpful healthcare data assistant. Your job is to find doctor information from the provided dataset.

Format:
- Doctor Name: [Insert Name]
- Department: [Insert Department]
- Hospital: [Insert Hospital]
- Address: [Insert Address]

Rules:
1. Only use the dataset below.
2. If multiple doctors match, list all.
3. If nothing found, say: "No matching doctor found."
4. Understand the user question and extract relevant information from the dataset to answer it.
5. Search doctor department based on the user question keywords, for example: "kidney" should match "nephrology".

Dataset:
{data}

User question:
{question}

Answer:
"""

prompt = ChatPromptTemplate.from_template(template)
chain = prompt | llm

# 4. Run
if __name__ == "__main__":
    csv_file = "doctors_data.csv"  # your saved CSV file
    
    print("Loading CSV data...")
    data_text = load_csv(csv_file)
    
    user_question = "Is there any doctor for child?"
    
    print("Searching...\n")
    response = chain.invoke({
        "question": user_question,
        "data": data_text
    })
    
    print("--- ASSISTANT RESPONSE ---")
    print(response.content)
```
